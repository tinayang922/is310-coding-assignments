## National Parks EDA

### This is an in-class excercise. A great opportunity to practice data exploration. 

```python
# import library
import pandas as pd
```

```python
# load the dataset
np_data = pd.read_csv("https://raw.githubusercontent.com/melaniewalsh/Neat-Datasets/main/1979-2020-National-Park-Visits-By-State.csv")
np_data
```

### In-class Exercise

```python
# Q1. What is the average number of visits for each state?
average_visits_state = np_data.groupby('State')['RecreationVisits'].mean().reset_index()
average_visits_state
```

```python
# Q2. What is the average number of visits for each National Park?
average_visits_park = np_data.groupby('ParkName')['RecreationVisits'].mean().reset_index()
average_visits_park
```

```python
# Q3. How many National Parks are there in each state?
park_counts = np_data.groupby('State')['ParkName'].count().reset_index()
park_counts
```
